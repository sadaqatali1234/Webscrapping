```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url="https://www.zong.com.pk/"
```

```python
res=requests.get(url)
```

```python
res
```

```python
res.content
```

```python
soup=BeautifulSoup(res.content,"html")
```

```python
soup.prettify
```

```python
paras=soup.find_all("p")
```

```python
paras
```

```python
anchors=soup.find_all("a")
```

```python
all_links=set()
for link in anchors:
  if (link.get("href")!="#"):
    all_links.add(link.get("href"))
```

```python
all_links
```

```python
df=pd.DataFrame(all_links)
```

```python
df.rename(columns={0:"Links"},inplace=True)
```

```python
df.iloc[8]
```

```python
df.to_excel("zong_link.xlsx")
```

```python
res_1.content
```

```python
soup=BeautifulSoup(res_1.content,"html")
```

```python
soup.prettify
```

```python
table=soup.find_all("table")
```

```python
len(table)
```

```python
df=pd.read_html(str(table))
```

```python
len(df)
```

```python
df_1=pd.DataFrame(df[0])
```

```python
df_1.head()
```

```python
df_1.drop(columns=["S.No.","Picture"],inplace=True)
```
